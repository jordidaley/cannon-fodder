```python
import ee
import geemap
import collections
collections.Callable = collections.abc.Callable
ee.Initialize()
import datetime
%matplotlib inline
import matplotlib.pyplot as plt
```

```python
Map = geemap.Map()

terraclimate = ee.ImageCollection("IDAHO_EPSCOR/TERRACLIMATE")
geometry = ee.Geometry.Point([77.54849920033682, 12.91215102400037])

# Assignment
# Use Gridded Climate dataset to chart a 40+ year time series
# if temparature at any location

# Workflow
# Load the TerraClimate collection
# Select the 'tmmx' band
# Scale the band values
# Filter the scaled collection
# Use ui.Chart.image.series() function to create the chart

# Hint1
# Data needed to be scaled by 0.1
# map() a function and multiply each image
# Multiplying creates a new image that doesn't have the same properties
# Use copyProperties() function to copy timestamp to new image
tmax = terraclimate.select('tmmx')

filtered = tmax \
  .filter(ee.Filter.date('1982-01-01', '2022-01-01')) \
  .filter(ee.Filter.bounds(geometry))

def func_gdc(image):
    return image.multiply(0.1) \
        .copyProperties(image,['system:time_start'])

tmaxScaled = filtered.map(func_gdc)

# Hint2
# You will need to specify a scale in meters for charting
# Use projection().nominalScale() to find the
# image resolution in meters
image = ee.Image(filtered.first())
print(image.projection().nominalScale().getInfo())
```

```python
def extract_data(image):
    stats = image.reduceRegion(**{ 
        'reducer':ee.Reducer.max(),
        'geometry':geometry,
        'scale':20
    })
    properties = {
        'month': image.date(),
        'tmmx': stats.get('tmmx')
    }
    return ee.Feature(None, properties)

data = ee.FeatureCollection(tmaxScaled.map(extract_data))
df = geemap.ee_to_pandas(data)
```

```python
dft = df
for i in range(len(df['month'])):
    dft['month'][i] = datetime.date.fromtimestamp(df['month'][i]['value']/1000.0)

dft.dropna(inplace=True)
dft.reset_index(drop=True, inplace=True)
```

```python
fig, ax = plt.subplots()
fig.set_size_inches(18,10)


dft.plot(ax=ax,
        title='Max Monthly Temperature at Bangalore',
        x='month',
        xlabel='Month',
        ylabel='Temperture (C)',
        kind='line',
        color='blue')
plt.tight_layout()
```
