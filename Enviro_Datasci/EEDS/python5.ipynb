```python
df2 = pd.read_csv("http://www.ldeo.columbia.edu/~rpa/usgs_earthquakes_2014.csv", infer_datetime_format=True, index_col="id")
```

```python
df2['time'] = pd.to_datetime(df2['time'], utc=True)
df2['updated'] = pd.to_datetime(df2['updated'], utc=True)
```

```python
df2.info()
```

```python
df2.describe()
```

```python
s = pd.Series(df2['mag'])
s.nlargest(20, keep='first')
```

```python
df2['country'] = pd.Series(df2['place']).str.split(",").str.get(1)
```

```python
df2.head()
```

```python
pd.Series(df2['country']).unique()
```

```python
dff = df2[df2['mag']>=4]
dff.size
```

```python
df5 = dff['country'].value_counts()
ax = df5.iloc[:5].plot.bar(x='country', y='count', rot=0)
plt.xlabel("country")
plt.ylabel("eq count")
```

```python
fig, (ax1, ax2) = plt.subplots(2, 1, figsize=(16,16))

df2.hist(column='mag', ax=ax1)
ax1.set_xscale('log')
dff.hist(column='mag', ax=ax2)
ax2.set_xscale('log')
```

```python
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(25, 8))

df2.plot.scatter(x='longitude',y='latitude',c='mag',cmap='viridis', ax=ax1)
dff.plot.scatter(x='longitude',y='latitude',c='mag',cmap='viridis', ax=ax2)
```
