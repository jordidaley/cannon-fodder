```python
data.toa_net_all_mon.mean()
```

```python
weight = data.toa_net_all_mon.lat
```

```python
weights = np.cos(np.radians(weight))
```

```python
pre_w = data.toa_net_all_mon
```

```python
weighted_pred = pre_w.weighted(weights)
```

```python
weighted_pred.mean()
```

```python
solar_w = data.solar_mon.weighted(weights)
sw_w = data.toa_sw_all_mon.weighted(weights)
lw_w = data.toa_lw_all_mon.weighted(weights)
```

```python
solar_w.mean(), sw_w.mean(), lw_w.mean()
```

```python
𝑅 = 6371
```

```python
dt = np.radians(1.0)
```

```python
datanot = data.toa_net_all_mon.mean(dim='time', keep_attrs=True)
sim = xr.ones_like(datanot)
```

```python
sima = sim * np.square(𝑅)*np.cos(np.radians(sim.lat))*dt*dt
```

```python
sima
```

```python
sima.sum()
```

```python
4*np.pi*np.square(𝑅)
```

```python
tnr = sima * datanot
```

```python
tnr
```

```python
tnr_lat = tnr.sum(dim='lon')
```

```python
tnr_lat.plot()
```

```python
tnr_lat.cumsum(dim='lat').plot()
```
